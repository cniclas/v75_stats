```python
from py.data_loader import DataLoader

# Test data
filepath = "data/data_v75_Romme.csv"

data_loader = DataLoader()  # Create a DataLoader instance
data_loader.load_data(filepath)  # Load the data into the DataLoader instance

df = data_loader.get_data() 
df.describe()
```

```python
from py.vector_input import VectorInput

test_vector_instance = VectorInput('Insatsprocent', 7)

test_vector_instance.sum_filter_options['min_sum'] = 0
test_vector_instance.sum_filter_options['max_sum'] = 25
test_vector_instance.sum_filter_options['in_nr_races'] = 2
test_vector_instance.sum_filter_options['selected_elements'] = [2,3,4]

data_vector = df['Insatsprocent']
print(data_vector)

td = test_vector_instance.get_relevant_column(df)
relevant_idxs = [i - 1 for i in test_vector_instance.sum_filter_options['selected_elements']]

relevant_data = test_vector_instance.get_relevant_elements(td[0], relevant_idxs)
print(relevant_data)

print(test_vector_instance.calculate_sums(relevant_data, test_vector_instance.sum_filter_options['in_nr_races']))

sum_filter_iloc = test_vector_instance.sum_filter_iloc(df)
print(sum_filter_iloc)
```
